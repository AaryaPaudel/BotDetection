```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
df = pd.read_csv('../data/fake reviews dataset.csv')
```

```python
print(df['label'].value_counts())
df['label'].value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Distribution of Fake (CG) vs Original (OR) Reviews')
plt.show()
```

```python
sns.countplot(data=df, x='rating', hue='label')
plt.title('Ratings Distribution by Label')
plt.show()
```

```python
analyzer = SentimentIntensityAnalyzer()
sample_text = df.iloc[0]['text_']
print(f"Sample Text: {sample_text}")
print(f"VADER Score: {analyzer.polarity_scores(sample_text)}")
```
